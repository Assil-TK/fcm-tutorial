```python
from fcmpy import FcmSimulator, FcmIntervention
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Weight matrix for Study Habits FCM
weight_matrix = pd.DataFrame(
    np.array([
        [0, 0, -0.4, 0.8],   # Hours_Studied effects
        [0, 0, -0.6, 0.5],   # Sleep_Quality effects
        [0, 0, 0, -0.7],     # Stress effects
        [0, 0, 0, 0]         # Exam_Score does not affect others
    ]),
    columns=["Hours_Studied", "Sleep_Quality", "Stress", "Exam_Score"]
)
weight_matrix
```

```python
# Initial state values
init_state = {
    "Hours_Studied": 0.6,
    "Sleep_Quality": 0.7,
    "Stress": 0.5,
    "Exam_Score": 0.0
}
init_state
```

```python
# Initialize the FCM simulator
sim = FcmSimulator()
```

```python
# Run simulation using sigmoid activation function
res_sigmoid = sim.simulate(
    initial_state=init_state,
    weight_matrix=weight_matrix,
    transfer='sigmoid',
    inference='mKosko',
    thresh=0.001,
    iterations=50,
    l=1
)
```

```python
#Plots
res_sigmoid.plot(marker='o')
plt.title("Study Habits FCM - Sigmoid Activation")
plt.xlabel("Iteration")
plt.ylabel("Concept Values")
plt.show()
```

```python
#Initialize interventions
inter = FcmIntervention(FcmSimulator)
inter.initialize(
    initial_state=init_state,
    weight_matrix=weight_matrix,
    transfer="sigmoid",
    inference="mKosko",
    thresh=0.001,
    iterations=50,
    l=1
)

# Add an intervention: Student uses a study planner to reduce stress
inter.add_intervention(
    "Study_Planner",
    type="continuous",
    impact={"Stress": -0.2},
    effectiveness=1
)

# Test intervention
inter.test_intervention("Study_Planner", iterations=50)
```

```python
#Plots
fig, axes = plt.subplots(1, 2, figsize=(12,5))

# Without intervention
res_sigmoid.plot(ax=axes[0], marker='o')
axes[0].set_title("Without Planner")
axes[0].set_xlabel("Iteration")
axes[0].set_ylabel("Concept Values")

# With intervention
inter.test_results["Study_Planner"].plot(ax=axes[1], marker='o')
axes[1].set_title("With Planner")
axes[1].set_xlabel("Iteration")

plt.tight_layout()
plt.show()
```
